### Experiment-1
Write a Python program to Create a NumPy array using a Python list, Display the array and print the shape and data type of the array.

```python
import numpy as np

# Create a NumPy array using a Python list
my_list = [10, 20, 30, 40, 50]
my_array = np.array(my_list)

# Display the array
print("Array:", my_array)

# Print the shape
print("Shape:", my_array.shape)

# Print the data type
print("Data Type:", my_array.dtype)
```

### Experiment-2
Write a program to Create a 3x4 matrix of zeros, create a 2x5 matrix of ones and print both matrices and their shapes.

```python
# Create a 3x4 matrix of zeros
zeros_matrix = np.zeros((3, 4))
print("3x4 Matrix of Zeros:\n", zeros_matrix)
print("Shape:", zeros_matrix.shape)
print()

# Create a 2x5 matrix of ones
ones_matrix = np.ones((2, 5))
print("2x5 Matrix of Ones:\n", ones_matrix)
print("Shape:", ones_matrix.shape)
```

### Experiment-3
Write a program to Create a NumPy array of numbers from 1 to 12, Reshape the array into a 3x4 matrix and display the original and reshaped array along with their shapes.

```python
# Create a NumPy array of numbers from 1 to 12
arr = np.arange(1, 13)
print("Original Array:", arr)
print("Original Shape:", arr.shape)
print()

# Reshape the array into a 3x4 matrix
reshaped_arr = arr.reshape(3, 4)
print("Reshaped Array (3x4):\n", reshaped_arr)
print("Reshaped Shape:", reshaped_arr.shape)
```

### Experiment-4
Write a program to create a NumPy array of 10 random integers, Find the minimum, maximum, mean, and sum of the array.

```python
# Create a NumPy array of 10 random integers
random_ints = np.random.randint(1, 100, size=10)
print("Random Array:", random_ints)

# Find minimum, maximum, mean, and sum
print("Minimum:", random_ints.min())
print("Maximum:", random_ints.max())
print("Mean:", random_ints.mean())
print("Sum:", random_ints.sum())
```

### Experiment-5
Write a Python program to create a Pandas DataFrame using a dictionary containing student names and marks. Display the DataFrame and print the data types (dtypes) of each column.

```python
import pandas as pd

# Create a dictionary containing student names and marks
data = {
    'Student_Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Marks': [85, 92, 78, 88, 95]
}

# Create a Pandas DataFrame using the dictionary
df = pd.DataFrame(data)

# Display the DataFrame
print("DataFrame:\n", df)
print()

# Print the data types (dtypes) of each column
print("Data Types:\n", df.dtypes)
```

### Preparing Sample Data for Experiments 6, 7, and 8

```python
# Let's create a sample CSV file to use in the following experiments
csv_data = '''ID,Name,Age,Score
1,John Doe,20,85.5
2,Jane Smith,22,90.0
3,Emily Davis,21,
4,Michael Brown,23,88.0
5,William Johnson,20,76.5
6,Emma Wilson,22,92.5
7,Oliver Martinez,21,80.0
8,Sophia Anderson,,85.0
9,James Taylor,23,78.0
10,Mia Thomas,20,95.0'''

with open('students_data.csv', 'w') as f:
    f.write(csv_data)
    
print("Sample dataset 'students_data.csv' created.")
```

### Experiment-6
Write a program to read a CSV file using read_csv(). Display the first 5 records using head() and display the last 5 records using tail() method.

```python
# Read the CSV file using read_csv()
df_csv = pd.read_csv('students_data.csv')

# Display the first 5 records using head()
print("First 5 records:")
display(df_csv.head())

# Display the last 5 records using tail()
print("\nLast 5 records:")
display(df_csv.tail())
```

### Experiment-7
Write a Python program to Load a dataset using Pandas Display dataset structure using info() and identify the number of rows, columns, and missing values.

```python
# Display dataset structure using info()
print("Dataset Structure Info:")
df_csv.info()

# Identify the number of rows and columns
rows, cols = df_csv.shape
print(f"\nNumber of rows: {rows}")
print(f"Number of columns: {cols}")

# Identify missing values
print("\nMissing values per column:\n", df_csv.isnull().sum())
```

### Experiment-8
Write a program to Read a CSV file into a DataFrame Display statistical summary using describe() Interpret mean, standard deviation, minimum, and maximum values.

```python
# Display statistical summary using describe()
print("Statistical Summary:")
stats = df_csv.describe()
display(stats)

print("\nInterpretation:")
print(f"- Mean: The average age is {stats.loc['mean', 'Age']:.2f} and the average score is {stats.loc['mean', 'Score']:.2f}.")
print(f"- Standard Deviation: The age varies by ~{stats.loc['std', 'Age']:.2f} years, and scores vary by ~{stats.loc['std', 'Score']:.2f}.")
print(f"- Minimum: The minimum age is {stats.loc['min', 'Age']:.0f} and the minimum score is {stats.loc['min', 'Score']:.1f}.")
print(f"- Maximum: The maximum age is {stats.loc['max', 'Age']:.0f} and the maximum score is {stats.loc['max', 'Score']:.1f}.")
```
